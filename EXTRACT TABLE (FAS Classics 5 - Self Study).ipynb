## IV. Data Integration and Transformation

```python
# Instead of using an if-else statement, we use np.where which have the same function. If a particular string contained in each particular row,
# we will enter that particular string
column = df['Column1']
df['DLE CATEGORY'] = np.where(column.str.contains("Depth and Breadth of Knowledge"), "Depth and Breadth of Knowledge",
                   np.where(column.str.contains("Research and Scholarship"), "Research and Scholarship",
                   np.where(column.str.contains("Level of Application of Knowledge"), "Level of Application of Knowledge", 
                   np.where(column.str.contains("Professional Capacity/ Autonomy"), "Professional Capacity/Autonomy",
                   np.where(column.str.contains("Level of Communications Skills"), "Level of Communications Skills",
                   np.where(column.str.contains("Awareness of Limits of Knowledge"), "Awareness of Limits of Knowledge"," " ))))))                             
# Make new DLE Category column if it contains one of the specific string above
df
```

```python
# We will make a new column which are the information in column3. We will use a split function after the string 'are:'
df['Column']= df['Column3'].str.split('are:').str[-1]
df.head()
```

```python
# For the next column, we will try using an if-else statement. Notice that this integration did not worked in our dataframe.
# Try with for loop with if else statement
for word in df['Column1']:
    if word == 'Knowledge':
        df['DLE CATEGORY DESCRIPTION'] = df['Column1'].str.split('Knowledge').str[-1] #the loop stops here
    elif word == 'Scholarship':
        df['DLE CATEGORY DESCRIPTION'] = df['Column1'].str.split('Scholarship').str[-1]
    else:
        df['DLE CATEGORY DESCRIPTION'] = df['Column1'].str.split('Autonomy').str[-1]
df
# didnt't work since for loop stopped when the first integration is fulfilled
```

### Since we can't use an if-else statement like above, we will instead use a replace function. For this column (DLE Category Description) , we will only require the string from Column1 after every DLE CATEGORY. We replace all the last DLE CATEGORY word with 'knowledge' before we split the column1 with the word 'Knowledge'.

```python
# Create a dictionary to replace all the word required with 'Knowledge'
replacement = {'Scholarship': 'Knowledge', 'Autonomy': 'Knowledge', 'Skills': 'Knowledge'}
df['Column1'] = df['Column1'].replace(replacement, regex = True)
df
```

```python
# Create a new column by splitting column1 with the string 'Knowledge'
df['DLE CATEGORY DESCRIPTION2'] = df['Column1'].str.split('Knowledge').str[-1]
df
```

```python
#The column PLO is the same as Column2, so we just need to create the same thing without any transformation
df['PLO'] = df['Column2']
df.head()
```

```python
#Split the PLO columns into multiple rows based on their DLE CATEGORY
replacement = {'2.': '1.', '3.': '1.', '4.': '1.', '5.': '1.'}
df['PLO'] = df['PLO'].replace(replacement, regex = True)
df['PLO'] = df['PLO'].str.split('1.')
df = df.apply (pd.Series.explode)
df
```

```python
# After we managed to get all the necessary columns, we will drop all the columns that won't be needed in our final result
df = df.drop(columns = ['Column1', 'Column2', 'Column3', 'DLE CATEGORY DESCRIPTION'])
df.head()
```

```python
# Rename the columns
df = df.rename(columns = {'DLE CATEGORY DESCRIPTION2':'DLE CATEGORY DESCRIPTION'})
df.head()
```
